Step 1: Import Required Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Step 2: Load the Iris Dataset

```python
# Load directly from seaborn's built-in datasets
df = sns.load_dataset('iris')
```

 Step 3: Display Dataset Structure

```python
print("Shape of dataset:", df.shape)        # Shape of Dataset
print("\nColumns:\n", df.columns)           # Column names
print("\nFirst 5 Rows:\n", df.head())       # Preview of first five rows
```

Step 4: Scatter Plot â€“ Relationships between Features

```python
# Scatter plot using seaborn pairplot
sns.pairplot(df, hue='species', palette='Set2')
plt.suptitle('Scatter Plot of Iris Features', y=1.02)
plt.show()
```

Step 5: Histogram â€“ Distribution of Each Feature

```python
df.hist(figsize=(8, 6), bins=15, color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Iris Features')
plt.tight_layout()
plt.show()
```

Step 6: Box Plot â€“ Outliers and Value Spread

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, orient="h", palette="Set3")
plt.title('Box Plots of Iris Features')
plt.show()
```
